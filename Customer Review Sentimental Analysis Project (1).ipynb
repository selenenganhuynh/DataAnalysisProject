```python
#to answer those questions and come up with some predictions
def remove_punctuation(text):
    final = "".join(p for p in text if p not in ("?", ".", ";", ":",  "!",'"'))
    return final
df['comments'] = df['comments'].apply(remove_punctuation)
df['title'] = df['title'].apply(remove_punctuation)
```

```python
#Create a new dataframe with the only information needed
dfNew = df[['sentiment','title','comments']]
dfNew.head()
```

```python
#Split the data into train and test set: 80% and 20%
#random split train and test data
index = df.index
df['random_number'] = np.random.randn(len(index))
train = df[df['random_number'] <= 0.8]
test = df[df['random_number'] > 0.8]
```

```python
#use CountVectorizer to transform text into vector with the frequency
#use for logistic regression later because it cannot understand text
#\b\w+\b: any text w is for word \b for the beginning and ending
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(token_pattern=r'\b\w+\b')
train_matrix = vectorizer.fit_transform(train['title'])
test_matrix = vectorizer.transform(test['title'])
```

```python
#import logistic regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
#Split targer and independent variables
X_train = train_matrix
X_test = test_matrix
y_train = train['sentiment']
y_test = test['sentiment']
```

```python
#fit logistic regression model on data
lr.fit(X_train,y_train)
```

```python
#makepredictions
predictions = lr.predict(X_test)
```

```python
#Testing
#find accuracy, precision, and recall:
from sklearn.metrics import confusion_matrix,classification_report
new = np.asarray(y_test)
confusion_matrix(predictions,y_test)
```

```python
print(classification_report(predictions,y_test))
```
